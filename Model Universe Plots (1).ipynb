```python
#matter curvature 
import math
import numpy as np
from matplotlib import pyplot as plt

# Simpson's 1/3 rule
def simpson(f, a, b, n):
    h = (b - a) / n
    s = f(a) + f(b)
    
    for i in range(1, n, 2):
        s += 4 * f(a + i * h)
    
    for i in range(2, n-1, 2):
        s += 2 * f(a + i * h)
    
    return s * h / 3

# Define the expressions as functions
exp = lambda x: 1 / math.sqrt((1.3 / x) - 0.3)
exp_2 = lambda x: 1 / math.sqrt((0.31 / x) + 0.69)

# Arrays for Big Crunch
x_11 = []
y_11 = []
i = 1.001
while i < 4.332:
    x_11.append(simpson(exp, 1, i, 1000))
    y_11.append(i)
    i += 0.001

x_12 = []
y_12 = []
i = 0.999
while i > 0:
    x_12.append(simpson(exp, 1, i, 1000))
    y_12.append(i)
    i -= 0.001

x_13 = []
y_13 = []
i = 4.332
while i > 0:
    x_13.append(simpson(exp, i, 4.332, 1000) + x_11[-1])
    y_13.append(i)
    i -= 0.001

x_values_1 = x_12[::-1] + x_11 + x_13
y_values_1 = y_12[::-1] + y_11 + y_13
t_values_1 = [13.6 * (value - x_values_1[0]) for value in x_values_1]

# Arrays for Big Chill
x_21 = []
y_21 = []
i = 1.001
while i < 10:
    x_21.append(simpson(exp_2, 1, i, 1000))
    y_21.append(i)
    i += 0.001

x_22 = []
y_22 = []
i = 0.999
while i > 0:
    x_22.append(simpson(exp_2, 1, i, 1000))
    y_22.append(i)
    i -= 0.001

x_values_2 = x_22[::-1] + x_21
y_values_2 = y_22[::-1] + y_21
t_values_2 = [13.7 * (value - x_values_2[0]) for value in x_values_2]

# Plotting                    
  
plt.figure(figsize=(10, 7))
plt.plot(t_values_1, y_values_1, label="Big Crunch, $\Omega_{m,0} = 1.30, k = +1$")
plt.plot(t_values_2, y_values_2, label="Big Chill, $\Omega_{m,0} = 0.31, k = -1$")
plt.xlabel("$t$ (Billion years)")
plt.ylabel("a")
plt.legend(loc="lower right")
plt.grid()
plt.show()
```

```python
#curvuture radiation                                        
  
import math
import numpy as np
from matplotlib import pyplot as plt

# Simpson's 1/3 rule implementation
def simpson(f, a, b, n):
    h = (b - a) / n
    x = np.linspace(a, b, n+1)
    fx = eval(f)
    k = fx[0] + fx[-1]
    k += 4 * np.sum(fx[1:-1:2])
    k += 2 * np.sum(fx[2:-2:2])
    return (h/3) * k

# Define expressions for different cosmological models
exp1 = "1/np.sqrt((0.69) + 0.31/(x*x))"
exp2 = "1/np.sqrt((0.31) + 0.69/(x*x))"
exp3 = "1/np.sqrt((-0.31) + 1.31/(x*x))"
exp4 = "1/np.sqrt((-0.69) + 1.69/(x*x))"
exp5 = "1/np.sqrt(1/(x*x))"

# Function to compute x and y values using Simpson's rule
def compute_values(exp, a, b, step_size):
    x_values = []
    y_values = []
    i = a
    while i <= b:
        try:
            integral_value = simpson(exp, 1, i, 1000)
            x_values.append(integral_value)
            y_values.append(i)
        except ValueError:
            break  # Skip if the expression leads to a math domain error
        i += step_size
    return x_values, y_values

# Compute values for each model
x_values_1, y_values_1 = compute_values(exp1, 1.001, 6, 0.001)
x_values_2, y_values_2 = compute_values(exp2, 1.001, 6, 0.001)
x_values_3, y_values_3 = compute_values(exp3, 1.001, 2.055, 0.001)
x_values_4, y_values_4 = compute_values(exp4, 1.001, 1.565, 0.001)
x_values_5, y_values_5 = compute_values(exp5, 1.001, 6, 0.001)

# Convert to time values
t_values_1 = [14.389 * (value - x_values_1[0]) for value in x_values_1]
t_values_2 = [14.389 * (value - x_values_2[0]) for value in x_values_2]
t_values_3 = [14.389 * (value - x_values_3[0]) for value in x_values_3]
t_values_4 = [14.389 * (value - x_values_4[0]) for value in x_values_4]
t_values_5 = [14.389 * (value - x_values_5[0]) for value in x_values_5]

# Plot the results
plt.figure(figsize=(10,7))
plt.plot(t_values_1, y_values_1, label="Big Chill, $\\Omega_{r,0} = 0.31$, k = -1")
plt.plot(t_values_2, y_values_2, label="Big Chill, $\\Omega_{r,0} = 0.69$, k = -1")
plt.plot(t_values_3, y_values_3, label="Big Chill, $\\Omega_{r,0} = 1$, k = 0")
plt.plot(t_values_4, y_values_4, label="Big Crunch, $\\Omega_{r,0} = 1.31$, k = +1")
plt.plot(t_values_5, y_values_5, label="Big Crunch, $\\Omega_{r,0} = 1.69$, k = +1")
plt.rc('font', size=12)
plt.xlabel("$t$ (Billion years)")
plt.ylabel("a")
plt.xlim(0, 250)  # Setting x-axis limits
lgnd = plt.legend(loc="lower right", fontsize=10)
lgnd.legendHandles[0]._sizes = [30]
lgnd.legendHandles[1]._sizes = [30]
lgnd.legendHandles[2]._sizes = [30]
lgnd.legendHandles[3]._sizes = [30]
lgnd.legendHandles[4]._sizes = [30]
plt.grid()
plt.show()
```
